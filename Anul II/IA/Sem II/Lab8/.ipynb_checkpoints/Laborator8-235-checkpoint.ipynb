```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load data
dirPath = "C:\\Users\\corneliu.dumitru\\Desktop\\University\\Anul II\\IA\\Sem II\\Lab8\\3d-points\\"
X = np.loadtxt(dirPath + "x_train.txt")
y = np.loadtxt(dirPath + "y_train.txt", 'int32') 
X_test = np.loadtxt(dirPath +'x_test.txt')
y_test = np.loadtxt(dirPath +'y_test.txt', 'int32') 
print(X.shape)
print(y.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
#visualize data
from mpl_toolkits import mplot3d

def plot3d_data(X, y):
    ax = plt.axes(projection='3d')
    ax.scatter3D(X[y == -1, 0], X[y == -1, 1], X[y == -1, 2],'b');
    ax.scatter3D(X[y == 1, 0], X[y == 1, 1], X[y == 1, 2],'r'); 
    plt.show()
    
def plot3d_data_and_decision_function(X, y, W, b): 
    ax = plt.axes(projection='3d')
    # create x,y
    xx, yy = np.meshgrid(range(10), range(10))
    # calculate corresponding z
    # [x, y, z] * [coef1, coef2, coef3] + b = 0
    zz = (-W[0] * xx - W[1] * yy - b) / W[2]
    ax.plot_surface(xx, yy, zz, alpha=0.5) 
    ax.scatter3D(X[y == -1, 0], X[y == -1, 1], X[y == -1, 2],'b');
    ax.scatter3D(X[y == 1, 0], X[y == 1, 1], X[y == 1, 2],'r'); 
    plt.show()

plot3d_data(X, y)
plot3d_data(X_test, y_test)
```

```python
#exercitiul 1
from sklearn.linear_model import Perceptron # importul clasei
perceptron_model = Perceptron(fit_intercept=True, max_iter=200, \
                              shuffle=True, eta0=0.1, tol = 1e-5)

perceptron_model.fit(X, y)

print(perceptron_model.score(X, y))
print(perceptron_model.coef_)
print(perceptron_model.intercept_)
print(perceptron_model.n_iter_)
print(perceptron_model.score(X_test, y_test))
W = perceptron_model.coef_
b = perceptron_model.intercept_

plot3d_data_and_decision_function(X, y, np.squeeze(W), b)
plot3d_data_and_decision_function(X_test, y_test, np.squeeze(W), b)
```

# Movie Review dataset

```python
import re
import os
from sklearn.utils import shuffle
from sklearn import preprocessing

class Data_Loader:
    def __init__(self):
        dataset_path = "C:\\Users\\corneliu.dumitru\\Desktop\\University\\Anul II\\IA\\Sem II\\Lab8\\data\\txt_sentoken\\"
        self.positive_samples_path = dataset_path + "pos/"
        self.negative_samples_path = dataset_path + "neg/"
        
    # functie care primeste path-ul catre un folder si
    # citeste toate fisierele din el, salvand recenziile intr-o lista
    def read_folder(self,folder_path):
        reviews = []
        for file_name in os.listdir(folder_path):
            file_path = folder_path + file_name
            file = open(file_path, 'r')
            file_content = file.read()
            reviews.append(re.sub("[-.,;:!?\"\'\/()_*=`]", "", file_content).split())
        return reviews
        
    def build_dataset(self):
        positive_samples = self.read_folder(self.positive_samples_path)
        negative_samples = self.read_folder(self.negative_samples_path)
        
        # impartim setul de date in 50% pentru antrenare si 50% pentru test
        num_training_samples_per_class = int(0.5 * len(positive_samples))
        num_test_samples_per_class = int(0.5 * len(positive_samples))
        
        # in setul de antrenare salvam exemplele pozitive si negative cu indecsii 0:499
        train_data = positive_samples[0:num_training_samples_per_class] + \
            negative_samples[0:num_training_samples_per_class]

        # exemplele pozitive vor avea eticheta 1, iar cele negative -1
        train_labels = [1] * num_training_samples_per_class + \
                     [-1] * num_training_samples_per_class
      
        # in setul de test salvam exemplele pozitive si negative  
        test_data = positive_samples[num_training_samples_per_class:len(positive_samples)] + \
                  negative_samples[num_training_samples_per_class:len(negative_samples)]
        test_labels = [1] * num_test_samples_per_class + \
                    [-1] * num_test_samples_per_class

        # amestecam datele 
        self.train_data, self.train_labels = shuffle(train_data, train_labels)
        self.test_data, self.test_labels = shuffle(test_data, test_labels)

class Bag_of_Words():
    
    def __init__(self):
        self.dictionar = {}
        self.nr_cuvinte = 0
    
    def  build_vocabulary(self, data):
        for lista in data:
            for cuvant in lista:
                if cuvant not in self.dictionar.keys():
                    self.dictionar[cuvant] = self.nr_cuvinte
                    self.nr_cuvinte = self.nr_cuvinte + 1
    
    def get_features(self, data):
        features = np.zeros((len(data),self.nr_cuvinte))
        for idx_lista, lista in enumerate(data):
            for cuvant in lista:
                if cuvant in self.dictionar.keys():
                    id = self.dictionar[cuvant]
                    features[idx_lista,id] += 1
        return features
```

```python
data_loader = Data_Loader()
data_loader.build_dataset()

bow_model = Bag_of_Words()
bow_model.build_vocabulary(data_loader.train_data)
train_features = bow_model.get_features(data_loader.train_data)
# normalizare
scaler = preprocessing.StandardScaler()
scaler.fit(train_features)
train_features_norm = scaler.transform(train_features)

#sanity check
print(train_features_norm.shape)
print(train_features_norm.mean(axis = 0))
print(train_features_norm.std(axis = 0))

test_features = bow_model.get_features(data_loader.test_data)
# normalizare
test_features_norm = scaler.transform(test_features)
#sanity check
print(test_features_norm.shape)
print(test_features_norm.mean(axis = 0))
print(test_features_norm.std(axis = 0))
```

```python
#exercitiul 2
import timeit
from sklearn.neural_network import MLPClassifier # importul clasei 
mlp_classifier_model = MLPClassifier(hidden_layer_sizes=(30,20 ), activation='tanh', \
                                     solver='sgd', learning_rate='constant', learning_rate_init=0.01, \
                                     max_iter=200, shuffle=True, random_state=None, tol=0.0001, \
                                     momentum=0.0)
print('training...')
start_time = timeit.default_timer()
mlp_classifier_model = mlp_classifier_model.fit(train_features_norm, data_loader.train_labels) 
end_time = timeit.default_timer()
print('proccesing time: in seconds', (end_time - start_time))
print('testing...')
print('plain - acuratetea pe multimea de antrenare este ', mlp_classifier_model.score(train_features_norm, data_loader.train_labels))
print('acuratetea pe multimea de testare este ', mlp_classifier_model.score(test_features_norm, data_loader.test_labels))
print('functia de activare de pe ultimul strat este %s' % mlp_classifier_model.out_activation_ )
print('numarul de iteratii parcurse pana la convergenta %d' % mlp_classifier_model.n_iter_)
```
